# Dataset Statistics
This script collects statistcs for all DLCC datasets

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# set paths to all files:

datasets = ['synthetic_ontology20/', 
            'synthetic_ontology20-2/',
            'synthetic_ontology50/',
            'synthetic_ontology100/',
            'synthetic_ontology1000_DLCCv1/']

# needs configure:
path_to_dataset='../data/Generated Datasets/'

patterns = ["tc01", "tc02", "tc03", "tc04", "tc05", "tc06", "tc07", "tc08", "tc09", "tc10", "tc11", "tc12" ]

config = 'configuration.txt'
graph = 'graph.nt'
```

```python
frame = pd.DataFrame()
columns_i = []
columns_a = []
for data in datasets:
    assertions = []
    nr_Instances = []
    for p in patterns:
        graph_path = path_to_dataset + data +  p + '/synthetic_ontology/'+graph
        with open(graph_path,'r') as c:
            for idx,v in enumerate(c.readlines(), 1):
                count=idx
            assertions.append(count)
        c.close()
        conf_path = path_to_dataset + data + p + '/synthetic_ontology/'+config
        with open(conf_path,'r') as c:
            for idx,v in enumerate(c.readlines(), 1):
                if idx == 2:
                    l = v.removeprefix('# of Instances : ')
                    nr_Instances.append(int(l))
                    break
    col_1_name = data + '_assertions'
    col_2_name = data + '_instances'
    columns_i.append(col_2_name)
    columns_a.append(col_1_name)
    frame[col_1_name] = assertions
    frame[col_2_name] = nr_Instances
```

```python
frame
```

Rename index to respective construct

```python
frame['idx'] = patterns
frame = frame.set_index('idx', drop=True).rename_axis(None)
```

```python
frame.describe()
```

# Outlier identified
tc04 in DLCC v1 is a big outlier.

```python
fig, ax = plt.subplots()
frame.boxplot(column=columns_i, ax=ax, notch=True,  rot=45, fontsize=10)
ax.set_xticklabels(['C (20)', 'D (20)', 'B (50)' ,'A (100)', 'DLCC v1'])
ax.set_ylabel('# of instances')
plt.show()
plt.close()
```

```python
fig, ax = plt.subplots()
frame.boxplot(column=columns_a, ax=ax, notch=True,  rot=45, fontsize=10)
ax.set_xticklabels(['C (20)', 'D (20)', 'B (50)' ,'A (100)', 'DLCC v1'])
ax.set_ylabel('# of assertions')
plt.show()
plt.close()
```

## Remove outliers of tc04 on DLCCv1

Replace outliers with median.

```python
# fix outliers by replacing them with the median
outlier = frame.copy()
outlier.loc['tc04', columns_i[4]] = frame[columns_i[4]].median()
outlier.loc['tc04', columns_a[4]] = frame[columns_a[4]].median()
```
