```python
# SET ALL PATHS
# Test set Directory
test_set_directory = '../data/Generated Datasets/synthetic_ontology20/'
patterns = ['tc01','tc02','tc03','tc04','tc05','tc06','tc07','tc08','tc09','tc10','tc11','tc12']
intermediate = '/synthetic_ontology/'
# change amount here
path_to_split = intermediate + '20/train_test/'

# results of DLCC Generator are misnamed; train and test are switched
# Train test not switched for 1000 instances 
test_file='train.txt'
namespace='http://example.org/'
graph_suffix ='_full.nt'
dlcc_config = 'configuration.txt'

# result directory
result_dir = '../data/DL-Learner_outputs/Ocel_rho/20-1/'
res_file_suffix = 'Res.txt'

# PATH TO RESULT to save heatmap later
result_fig = result_dir 
```

```python
import re 
def fix_format(string):
    return re.findall('(?<=<).*?(?=>)', string)[0]
```

```python
def retrieve_s_in_F(stri, file):
    with open(file, 'r') as f:
        for index, line in enumerate(f):
            # search string
            if stri in line:
                f.close()
                return line
        f.close()
    return 'not found'
```

```python
def str_in_f(string, file):
    with open(file) as f:
        if string in f.read():
            f.close()
            return True
    f.close()
    return False
```

```python
def generate_manchester_syntax(p, file):
    # Terminology derived from DLCC config files
    edge_str = 'Target edge'
    clazz_str = 'Target class'
    instance_str = 'Target instance'
    property_str = 'Target property'
    pos_c_str = 'Positive object class'
    pos_s_str = 'Positive subject class'
    tar_c_str = 'Target object class'
    tar_s_str = 'Target subject class'
    if p == patterns[0]:
        edge = retrieve_s_in_F(edge_str, file)
        edge = fix_format(edge)
        syntax = edge + " some Thing"
    elif p == patterns[1]:
        edge = retrieve_s_in_F(edge_str, file)
        edge = fix_format(edge)
        syntax = "inverse (" + edge + ") some Thing"
    elif p == patterns[2]:
        edge = retrieve_s_in_F(edge_str, file)
        edge = fix_format(edge)
        syntax = "("+edge+" some Thing) or (inverse ("+edge+") some Thing)"
    elif p == patterns[3]:
        instance = retrieve_s_in_F(instance_str, file)
        instance = fix_format(instance)
        print(instance)
        syntax = 'dont know'
    elif p == patterns[4]:
        instance = retrieve_s_in_F(instance_str, file)
        instance = fix_format(instance)
        print(instance)
        syntax = 'dont know'
    elif p == patterns[5]:
        edge = retrieve_s_in_F(edge_str, file)
        edge = fix_format(edge)
        instance = retrieve_s_in_F(instance_str, file)
        instance = fix_format(instance)
        syntax = edge + " some ({" + instance + "})"
    elif p == patterns[6]:
        prop = retrieve_s_in_F(property_str, file)
        prop = fix_format(prop)
        pos = retrieve_s_in_F(pos_c_str, file)
        pos = fix_format(pos)
        syntax = prop + " some " + pos 
    elif p == patterns[7]:
        prop = retrieve_s_in_F(property_str, file)
        prop = fix_format(prop)
        pos = retrieve_s_in_F(pos_s_str, file)
        pos = fix_format(pos)
        syntax = "inverse (" + prop + ") some " + pos 
    elif p == patterns[8]:
        prop = retrieve_s_in_F(property_str, file)
        prop = fix_format(prop)
        syntax = prop + " min 2 Thing"
    elif p == patterns[9]:
        prop = retrieve_s_in_F(property_str, file)
        prop = fix_format(prop)
        syntax = "inverse (" + prop + ") min 2 Thing"
    elif p == patterns[10]:
        prop = retrieve_s_in_F(property_str, file)
        prop = fix_format(prop)
        obj_c = retrieve_s_in_F(tar_c_str, file)
        obj_c = fix_format(obj_c)
        syntax = prop + " min 2 " + obj_c
    elif p == patterns[11]:
        prop = retrieve_s_in_F(property_str, file)
        prop = fix_format(prop)
        obj_s = retrieve_s_in_F(tar_s_str, file)
        obj_s = fix_format(obj_s)
        syntax = "inverse (" + prop + ") min 2 " + obj_s
    return syntax
```

```python
for p in patterns:
    print('===================================')
    file_path= test_set_directory + p + intermediate + dlcc_config
    sy = generate_manchester_syntax(p, file_path)
    print(p)
    print(f'expected: \t{sy}')
    res_path = result_dir+p+res_file_suffix
    found=retrieve_s_in_F(sy, res_path)
    print(f'found: \t\t{found}')
    print(f'success? --> {str_in_f(sy, res_path)}') 
```
